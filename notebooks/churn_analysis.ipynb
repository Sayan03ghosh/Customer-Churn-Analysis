```python
from pathlib import Path
import pandas as pd

base = Path.cwd()
candidates = [
    base / 'data' / 'Telco Customer Churn Dataset.csv',
    base / 'data' / 'telco_churn.csv',
    base.parent / 'data' / 'Telco Customer Churn Dataset.csv',
    base.parent / 'data' / 'telco_churn.csv',
]

csv_path = next((p for p in candidates if p.exists()), None)
if csv_path is None:
    raise FileNotFoundError('Could not find churn CSV in expected data paths.')

df = pd.read_csv(csv_path)
display(df.head())
print(f'Loaded: {csv_path}')
```

```python
# Avoid chained inplace operations (pandas 3.0-safe)
df['Offer'] = df['Offer'].fillna('No Offer')
df['Internet Type'] = df['Internet Type'].fillna('No Internet')
df['Churn Category'] = df['Churn Category'].fillna('No Churn')
df['Churn Reason'] = df['Churn Reason'].fillna('No Churn')

df[['Offer', 'Internet Type', 'Churn Category', 'Churn Reason']].isna().sum()
```
